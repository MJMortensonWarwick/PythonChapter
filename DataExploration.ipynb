```python
### Working with data ###

import pandas as pd

#load file from working directory
advertising = pd.read_csv("advertising.csv", sep = ",", 
                          encoding = "ISO-8859-1")

print(advertising.dtypes) # print data types

advertising.head() # first five rows

# count N/A values
advertising.isna().sum()
```

```python
### Histograms ###

import matplotlib.pyplot as plt 

# if using Jupyter Notebooks also include this line which
# will print graphs to screen
%matplotlib inline

# the variable n_bins will determine how many bins to
# include in the historgram
n_bins = 12

# create the histogram and set the axes labels/title
plt.hist(advertising['Sales'], n_bins, histtype='bar')
plt.xlabel('Sales volume')
plt.ylabel('Frequency')
plt.title('Histogram of Sales Volumes')
plt.show()
```

```python
### Descriptives and box plot ###

# Descriptive statistics
print(advertising['Sales'].describe())

# Box plot
import seaborn as sns
sns.set(style='whitegrid') 
ax = sns.boxplot(x=advertising['Sales'])
```

```python
### Scatter plots ###

import matplotlib.pyplot as plt

# if using Jupyter Notebooks also include this line which
# will print graphs to screen
%matplotlib inline

plt.scatter(advertising['TV'], advertising['Sales'], marker='o')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Scatter Plot of TV and Sales')
plt.show()
```

```python
### Scatter plot with trendline ###

import numpy as np
import matplotlib.pyplot as plt

# if using Jupyter Notebooks also include this line which
# will print graphs to screen
%matplotlib inline

plt.plot(advertising['TV'], advertising['Sales'],'o')
# calculate the trendline
z = np.polyfit(advertising['TV'], advertising['Sales'], 1)
p = np.poly1d(z)
plt.plot(advertising['TV'], p(advertising['TV']),"r--")
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Scatter Plot of TV and Sales')
plt.show()
```

```python
### Scatter matrix ###

import seaborn as sns
sns.set(style="ticks")
sns.pairplot(advertising[["TV", "Radio", "Press", "Sales"]])
```
